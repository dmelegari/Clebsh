```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import sqrt, symbols
from sympy.physics.quantum.cg import CG

# Lista per memorizzare i massimi
max_values = []

n_values = list(range(2, 10))

# Itera su tutti i valori di n
for n in n_values:
    j = (2**n - 1)/2

    J_values = [i for i in range(0, int(2*j) + 1)]

    m_values = [-j + i for i in range(int(2*j + 1))]

    max_sum_per_n = float('-inf')

    # Itera su tutti i valori di m
    for m in m_values: 
        vals = []
        for J in J_values:
            if J >= 2*m:
                cg_coeff_squared = float(CG(j, m, j, m, J, int(2*m)).doit())**2
                vals.append(cg_coeff_squared/(2*J+1))
        
        current_sum = np.sum(vals)
        max_sum_per_n = max(max_sum_per_n, current_sum)
    
    max_values.append(max_sum_per_n)
    print(f"For n={n}, Max Sum = {max_sum_per_n}")
```

```python
# Genera il grafico

curve = [2**(-n/2) for n in n_values]

plt.plot(n_values, max_values, marker='o')
plt.plot(n_values, curve, marker='o')

plt.title('Massimo della somma per ogni valore di n')
plt.xlabel('n')
plt.ylabel('Massimo della somma')
plt.yscale('log')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols
from math import factorial

n = 9
j = (2**n - 1) / 2

sum_values = []  # Lista per memorizzare i valori della somma

sum = 0.0  # Inizializza sum con 0.0 invece di 0

for J in range(int(2 * j) + 1):
    num = (factorial(int(2 * j)))**2
    den = factorial(J + int(2 * j) + 1) * factorial(int(2 * j) - J)
    coeff = num / den

    sum += coeff
    sum_values.append(sum)

    # Verifica il criterio solo quando sum è diverso da zero
    if sum != 0 and abs(coeff / sum) < 1e-8 and J > 0:
        
        print('n      = ', n)
        print('j      = ', j)
        print('J_stop = ', J)
        print('sum    = ', sum)
        break  # Interrompi il ciclo se il criterio è soddisfatto

# Creazione del grafico
J_values = np.arange(0, len(sum_values))
plt.plot(J_values, sum_values, marker='o')
plt.title('Accumulated Sum vs J')
plt.xlabel('J')
plt.ylabel('Accumulated Sum')
plt.grid(True)
plt.show()
```
